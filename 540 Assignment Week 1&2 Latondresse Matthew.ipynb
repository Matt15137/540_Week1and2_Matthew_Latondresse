```python
#Activity 1

import random

# Initialize a list to store the average differences
average_differences = []

# Repeat the experiment three times
for _ in range(3):
    # Step 1: Create a list of 100 random numbers
    random_numbers = [random.randint(1, 1000) for _ in range(100)]
    print(random_numbers)
    
    # Step 2: Create a new list with numbers divisible by 3
    divisible_by_three = [num for num in random_numbers if num % 3 == 0]
    print("List of numbers divisible by three:", divisible_by_three)
    
    # Step 3: Calculate the length of both lists and store the difference
    difference = len(random_numbers) - len(divisible_by_three)
    print("Difference", difference)
    
    # Step 4: Append the difference to the average differences list
    average_differences.append(difference)

# Step 5: Calculate the arithmetic mean of the three difference values
mean_difference = sum(average_differences) / len(average_differences)

print("Average differences:", average_differences)
print("Arithmetic mean of differences:", mean_difference)
```
